```
from scipy.optimize import minimize
import matplotlib.pyplot as plt
import numpy as np
import math

num_of_fractions = 41
t = np.linspace(0, 2, num_of_fractions) 
dt = t[1] - t[0]  
exponenta = [math.exp(2* i) for i in t]

def f(y):
    return np.sum(3 * y[1:] ** 2 + ((y[1:] - y[:-1]) / dt) ** 2 + ((y[1:] - y[:-1]) / dt) * y[1:] * 12 + 14 * y[1:] * exponenta[1:] ) # Здесь нужно поменять 
                                                                                                                                      # коэфиценты перед всеми слегаемыми

y0 = [0.5 for x in range(0, num_of_fractions)]  

total_model = 0
for i in range(100):
  y = np.random.normal(loc=10.0, scale=30.0)  # подставить свои значения, где loc-мат.ожидание scale-дисперсия

  bounds = [(None, None) for x in range(0, num_of_fractions)]
  bounds[0], bounds[-1] = (y, y), (19.0, 19.0) # Подствить своё значение y(2) вместо 19, то есть (y(2), y(2))

  res = minimize(f, y0, method='l-bfgs-b', bounds=bounds)

  total_model += res.x
plt.plot(t, total_model / 100)
plt.xlabel('t')
plt.ylabel('y')
plt.legend('y численный')
plt.show()
```
