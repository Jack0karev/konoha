```python
# Создаю пустой массив, чтобы его заполнить числами
import numpy as np
arr = np.zeros((df1.shape[0] * 5, len(cols)))
```

```python
# Заменяю все прочерки и NM на 999999999, т.к. массив может содержать только числа
df1 = df1.replace('-', 999999999)
df1 = df1.replace('NM', 999999999)
# Создаю пустой массив, чтобы его заполнить числами
import numpy as np
arr = np.zeros((df1.shape[0] * 5, len(cols)))

# Сам цикл
for i in range(2, len(cols)):
    counter = 0
    for j in df1['Company Name']:
        year = 2016
        for n in range(5*(i-2), 5*(i-1)):
            arr[counter][i] =  df1[value[n]][df1['Company Name'] == j].values[0]

            arr[counter][1] = year
            counter += 1
            year += 1
```

```python
# Записываю все значения в таблицу
tot = pd.DataFrame(arr,columns=cols)
# Преобразую года из числового формата в во временной
tot.Years = pd.to_datetime(tot.Years, format='%Y')
# Создаю таблицу для названия компаний
names = []
for i in df1['Company Name']:
    for j in range(5):
        names.append(i)
tot['Company Name'] = names
```

```python
# Доделывал отдельно с данными, где есть строки и цифры вместе
x = ['Salary  [FY 2016] ($USD, 5/2/2020 rate)',
       'Salary  [FY 2017] ($USD, 5/2/2020 rate)',
       'Salary  [FY 2018] ($USD, 5/2/2020 rate)',
       'Salary  [FY 2019] ($USD, 5/2/2020 rate)',
       'Total Annual Cash Compensation  [FY 2016] ($USD, 5/2/2020 rate)',
       'Total Annual Cash Compensation  [FY 2017] ($USD, 5/2/2020 rate)',
       'Total Annual Cash Compensation  [FY 2018] ($USD, 5/2/2020 rate)',
       'Total Annual Cash Compensation  [FY 2019] ($USD, 5/2/2020 rate)'
]
```

```python
import re
arr = np.zeros((df1.shape[0]*5, 2))
for i in range(2):
    counter = 0
    for j in df1['Company Name']:
        con = 1
        for n in range(4*(i), 4*(i+1)):
            s = df1[x[n]][df1['Company Name'] == j].values[0]
            s = re.sub('[A-z () ;’-]', ' ', s)
            if '’' in s: print(s)
            s = re.sub(',', '',s).split()
            s = [int(word) for word in s if word.isdigit()]
            np.mean(s)
            arr[counter][i] =  np.mean(s)
            counter += 1 
            con += 1
        arr[counter][i] =  999999999
        counter += 1
    
tot['Salary'] = arr[:, 0]
tot['Total annual Cash'] = arr[:, 1]
# Добавил число сотрудников
employees = []
for i in df1['Number of Employees - Global (Latest)']:
    for j in range(5):
        employees.append(i)
tot['Number of Employees'] = employees
# Вернул все пропуски к прочеркам
tot = tot.replace(999999999, '-')
tot.head()
```
